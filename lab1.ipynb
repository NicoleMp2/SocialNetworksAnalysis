```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
n = 44 # μέγεθος δικτύου
d = 4 
p = 0 # όχι τυχαίο δίκτυο
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG = nx.watts_strogatz_graph(n,d,p) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
rger_seed = random.seed(603) # επιλέγω seed λίγο μεγαλύτερο γιατί ο γράφος που παράγεται με το αρχικό δεν είνια fully connected
n = 44
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n,M,seed=rger_seed) 

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
s = random.seed(seedno)
n = 44
R = 0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n,R,seed= s)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html

pos = nx.get_node_attributes(RGG, "pos")
#find node near centre
dmin = 1 
ncentre = 0
for n in pos:
    x, y = pos[n]
    d = (x-0.5)**2 + (y-0.5)**2
    if d < dmin:
        ncentre = n
        dmin = d

path = dict(nx.single_source_shortest_path_length(RGG,ncentre))
plt.figure(figsize=(8,8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG, 
    pos, 
    nodelist=list(path.keys()),
    node_size=80,
    node_color=list(path.values()),
    cmap=plt.cm.Blues_r
)

plt.xlim(-0.05,1.05)
plt.ylim(-0.05,1.05)
plt.axis("off")
plt.show()
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n = 44
d=4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF = nx.barabasi_albert_graph(n, d) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n = 44
d=4
p=0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW = nx.watts_strogatz_graph(n,d,p) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW)
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):

#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.


```python
graphs = {'REG':REG,'RGER':RGER,'RGG':RGG,'SF':SF,'SW':SW}
```

```python
#returns a dictionary {node : clustering coefficient}
def cc(G):
    g = {key: round(val,4) for key, val in nx.clustering(G).items() }
    return g

#plot stuff
figure = plt.figure()
figure.subplots_adjust(hspace=0.4,wspace=0.4)
spec = matplotlib.gridspec.GridSpec(ncols=6, nrows=2) # 6 columns evenly divides both 2 & 3

ax1 = figure.add_subplot(spec[0,0:2]) 
ax2 = figure.add_subplot(spec[0,2:4])
ax3 = figure.add_subplot(spec[0,4:])
ax4 = figure.add_subplot(spec[1,1:3]) 
ax5 = figure.add_subplot(spec[1,3:5])



for name,graph in graphs.items():
    g = cc(graph)
    print("---------",name,"---------")
    print(g)
    if(name == 'REG'):
        ax1.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax1.set_title("Clustering Coefficient Distribution of REG",fontsize='xx-small')
    elif(name == 'RGER'):
        ax2.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax2.set_title("Clustering Coefficient Distribution of RGER",fontsize='xx-small')
    elif(name == 'RGG'):
        ax3.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax3.set_title("Clustering Coefficient Distribution of RGG",fontsize='xx-small')
    elif(name == 'SF'):
        ax4.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax4.set_title("Clustering Coefficient Distribution of SF",fontsize='xx-small')
    else:
        ax5.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax5.set_title("Clustering Coefficient Distribution of SW",fontsize='xx-small')
    


```

#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.


```python
def av_clustering_variance(g):
    av_clust = nx.average_clustering(g)
    v = np.array(list(cc(g).values()))
    v = np.var(v)
    return av_clust, v

for name,graph in graphs.items():
    av, var = av_clustering_variance(graph)
    print(name,"\t-> average clustering :",round(av,3), "\tvariance :", round(var,3))
```

#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 


```python
def min_max_clustering(g):
    gr = cc(g)
    mini = min(gr.items(), key=lambda x: x[1])
    maxi = max(gr.items(), key=lambda x: x[1])
    return mini, maxi

for name,graph in graphs.items():
    mini, maxi = min_max_clustering(graph)
    print(name,"\t-> node with minimum CC :",mini, "\tnode with maximum CC :", maxi)
```

#### Σχολιασμός αποτελεσμάτων


Ιδιαίτερο ενδιαφέρον παρουσιάζει η γραφική παράσταση όπου φαίνονται οι κατανομές των τιμών των clustering coefficients για τους κόμβους όλων των γράφων:
    - Όπως είναι αναμενόμενο, στον κανονικό γράφο, ο clustering coefficient είναι σταθερός για όλους τους κόμβους.
    - Στον random geometric graph, οι τιμές του clustering coefficient είναι σημαντικά μεγάλες (πάνω από 0.5 σε όλους τους κόμβους), κάτι που διαισθητικά βγάζει νόημα, αφού οι κόμβοι σχηματίζουν γειτονιές με βάση τις γεωμετρικές αποστάσεις τους.
    - Στον γράφο scale-free, όλοι οι κόμβοι έχουν σχετικά μικρό συντελεστή, αφού η πιθανότητα ένας κόμβος να έχει μεγάλο βαθμό (και άρα πολλούς γείτονες, και άρα μεγάλο συντελεστή) είναι εκθετικά μικρή.


### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.

```python
def compute_path_lengths(g):
    return nx.all_pairs_shortest_path_length(g) # αυτός είναι ένας generator 

# παίρνω μια λιστα από τουπλες
# κάθε τούπλα είναι της μορφης (int, dict)
# ο int είναι ο κόμβος τον οποίο θεωρούμε σαν αφετηρία κάθε φορά
# το dict είναι η απόσταση του κόμβου αφετηρίας από τον εκάστοτε κόμβο
# πχ για αφετηριά 0 {0: 0, 1:1, 2:1, 43: 1, 42: 1, 3: 2, 4: 2, 40: 2} ο κόμβος 0 απέχει από τον κόμβο3 2

def compute_average_path_length_variance(g):
    #return sum([sum(x[1].values()) for x in compute_path_lengths(g)])/(len(g.nodes)**2)
    nodes = len(g.nodes)
    list_mean = [sum(x[1].values())/(nodes) for x in compute_path_lengths(g)]
    mean = sum(list_mean) / nodes
    var = np.var(list_mean)
    return mean,var



for name,graph in graphs.items():
    av_path_length = compute_average_path_length_variance(graph)
    print(name,"\t-> average path length :",round(av_path_length[0],3), "\tvariance :", round(av_path_length[1],3))
```

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(g):
    #for each node find the degree of the node and its neighbors with min path lenght = diameter
    #return the nodes with the max degree and the neighbors
    nodes = []
    for node in g.nodes:
        neighbors =  g.neighbors(node)
        degree = g.degree[node]
        nodes.append((node,degree,list(neighbors)))
    maxi = max(nodes, key=lambda x: x[1])
    l = [x for x in nodes if x[1] == maxi[1]]
    return l
    
    
for name,graph in graphs.items():
    fg = find_nodes(graph)
    print("---------",name, "---------")
    for i in fg:
        print("node : ",i[0]," has max max degree of", i[1], "\tThe neighbors are : ", i[2])
```

```python
# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο. 
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    # return the periphery of the graph
    # the periphery is the set of nodes with eccentricity equal to the diameter
    return [n for n in G.nodes() if nx.eccentricity(G, v=n) == nx.diameter(G)]


for name,graph in graphs.items():
    print(name,"\t-> periphery :",find_periphery(graph))
```

```python
def find_center(G):
    #find the center of the graph
    #the center is the set of nodes with eccentricity equal to the radius
    return [n for n in G.nodes() if nx.eccentricity(G, v=n) == nx.radius(G)]


for name,graph in graphs.items():
    print(name,"\t-> center :",find_center(graph))
```

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)

##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.


```python
def find_degree(G):
    #find the degree of the graph
    return G.degree()


#plot stuff
figure = plt.figure()
figure.subplots_adjust(hspace=0.4,wspace=0.4)
spec = matplotlib.gridspec.GridSpec(ncols=6, nrows=2) # 6 columns evenly divides both 2 & 3

ax1 = figure.add_subplot(spec[0,0:2]) 
ax2 = figure.add_subplot(spec[0,2:4])
ax3 = figure.add_subplot(spec[0,4:])
ax4 = figure.add_subplot(spec[1,1:3]) 
ax5 = figure.add_subplot(spec[1,3:5])

#thelei allagi
for name,graph in graphs.items():
    g = find_degree(graph)
    if(name == 'REG'):
        ax1.hist(list(dict(graphs['REG'].degree).values()), bins=15, color="lightblue", ec="blue",lw=1)
        ax1.set_title("Degree Distribution of REG",fontsize='xx-small')
    elif(name == 'RGER'):
        ax2.hist(list(dict(graphs['RGER'].degree).values()), bins=15, color="lightblue", ec="blue",lw=1)
        ax2.set_title("Degree Distribution of RGER",fontsize='xx-small')
    elif(name == 'RGG'):
        ax3.hist(list(dict(graphs['RGG'].degree).values()), bins=15, color="lightblue", ec="blue",lw=1)
        ax3.set_title("Degree Distribution of RGG",fontsize='xx-small')
    elif(name == 'SF'):
        ax4.hist(list(dict(graphs['SF'].degree).values()), bins=15, color="lightblue", ec="blue",lw=1)
        ax4.set_title("Degree Distribution of SF",fontsize='xx-small')
    else:
        ax5.hist(list(dict(graphs['SW'].degree).values()), bins=15, color="lightblue", ec="blue",lw=1)
        ax5.set_title("Degree Distribution of SW",fontsize='xx-small')
    

```

##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.


```python

for name,graph in graphs.items():
    dn = find_degree(graph)
    #find mean value of dn
    mean = sum([x[1] for x in dn])/len(dn)
    #find variance of dn
    var = np.var([x[1] for x in dn])
    print(name,"\t-> mean degree :",round(mean,3), "\tvariance :", round(var,3))
```

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)

##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.


```python
def find_closeness(g):
    #find the closeness of the graph
    return nx.closeness_centrality(g)

#plot stuff
figure = plt.figure()
figure.subplots_adjust(hspace=0.4,wspace=0.4)
spec = matplotlib.gridspec.GridSpec(ncols=6, nrows=2) # 6 columns evenly divides both 2 & 3

ax1 = figure.add_subplot(spec[0,0:2]) 
ax2 = figure.add_subplot(spec[0,2:4])
ax3 = figure.add_subplot(spec[0,4:])
ax4 = figure.add_subplot(spec[1,1:3]) 
ax5 = figure.add_subplot(spec[1,3:5])



for name,graph in graphs.items():
    g = find_closeness(graph)
    if(name == 'REG'):
        ax1.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax1.set_title("Closeness Centrality Distribution of REG",fontsize='xx-small')
    elif(name == 'RGER'):
        ax2.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax2.set_title("Closeness Centrality Distribution of RGER",fontsize='xx-small')
    elif(name == 'RGG'):
        ax3.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax3.set_title("Closeness Centrality Distribution of RGG",fontsize='xx-small')
    elif(name == 'SF'):
        ax4.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax4.set_title("Closeness Centrality Distribution of SF",fontsize='xx-small')
    else:
        ax5.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax5.set_title("Closeness Centrality Distribution of SW",fontsize='xx-small')
    

```

##### Γ2.2 Να υπολογιστεί η μέση εγγύτητα καθώς και η διασπορά της για κάθε τοπολογία.


```python
#find mean value of closeness and variance
for name,graph in graphs.items():
    dn = find_closeness(graph)
    #find mean value of dn
    mean = sum(dn.values())/len(dn)
    #find variance of dn dh is a dictionary
    var = np.var(list(dn.values()))
    print(name,"\t-> mean closeness :",round(mean,3), "\tvariance :", round(var,3))
```

Η κεντρικότητα ενός κόμβου εκφράζει την "σημασία" του κόμβου αυτού για τον γράφο.
Παρατηρώ ότι στον κανονικό γράφο, όλοι οι κόμβοι έχουν την ίδια σημασία, όπως περιμέναμε λόγω της συμμετρίας του γράφου.

Ο RGG θυμίζει κι αυτός την κανονική κατανομή σε μικρότερο βαθμό.
Στον Small-World γράφο, η διασπορά της κεντρικότητας είναι από τις μικρότερες συγκριτικά με τους υπόλοιπους γράφους.
Οι κόμβοι με μεγαλύτερη σημασία πιθανώς να είναι οι κόμβοι όπου προσκύπτουν οι ακμές "συντομεύσεις".
Στον Scale-Free γράφο δεν παρατηρώ κάποια ομοιομορφία.

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.


```python
#import scipy.sparse.linalg as linalg
def find_katz(g):
    #find max eigenvalue of the adjacency matrix
    max_eig = max(nx.adjacency_spectrum(g))
    # the alpha is 1/max_eig-0.01
    alpha = 1/(max_eig-0.01)
    return nx.katz_centrality(g,alpha=alpha)

#plot stuff
figure = plt.figure()
figure.subplots_adjust(hspace=0.4,wspace=0.4)
spec = matplotlib.gridspec.GridSpec(ncols=6, nrows=2) # 6 columns evenly divides both 2 & 3

ax1 = figure.add_subplot(spec[0,0:2]) 
ax2 = figure.add_subplot(spec[0,2:4])
ax3 = figure.add_subplot(spec[0,4:])
ax4 = figure.add_subplot(spec[1,1:3]) 
ax5 = figure.add_subplot(spec[1,3:5])



for name,graph in graphs.items():
    g = find_katz(graph)
    print("---------",name,"---------")
    print(g)
    if(name == 'REG'):
        ax1.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax1.set_title("Clustering Coefficient Distribution of REG",fontsize='xx-small')
    elif(name == 'RGER'):
        ax2.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax2.set_title("Clustering Coefficient Distribution of RGER",fontsize='xx-small')
    elif(name == 'RGG'):
        ax3.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax3.set_title("Clustering Coefficient Distribution of RGG",fontsize='xx-small')
    elif(name == 'SF'):
        ax4.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax4.set_title("Clustering Coefficient Distribution of SF",fontsize='xx-small')
    else:
        ax5.hist(g.values(), bins=15, color="lightblue", ec="blue",lw=1)
        ax5.set_title("Clustering Coefficient Distribution of SW",fontsize='xx-small')
    

```
